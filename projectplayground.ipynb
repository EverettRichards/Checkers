```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm
import time
from GameState import GameState
```

```python
def playRandomGame(max_turns=10,wait_time_ms=1000):
    myboard = GameState()
    myboard.plot()
    p = 1
    turn = 1
    while True:
        # Get a list of legal moves
        legal_moves = myboard.getLegalMoves(p)

        # Out of luck? End the game!
        if len(legal_moves)==0 or turn > max_turns:
            break

        # Select one of the legal moves that the player can make
        move = legal_moves[0]
        if len(legal_moves) > 1:
            move = legal_moves[np.random.randint(0,len(legal_moves)-1)]

        # Actually execute the move
        myboard.performMove(move)

        # Visually plot the results of the move
        myboard.plot(wait_time_ms=wait_time_ms,last_move=move[1])

        # Prepare for the next loop iteration
        p = -p # switch players :)
        turn += 1
    print(f"Ended after {turn-1} turns")
    print(myboard)
```

```python
playRandomGame(max_turns=100,wait_time_ms=50)
```
